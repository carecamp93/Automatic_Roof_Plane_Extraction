```python
import numpy as np
import pickle
from numpy import array
import matplotlib.pyplot as plt
import matplotlib.image as image
```

```python
# Define the input annot files
input_annot ="2. TrainingPreparation/Enschede3D/annot"

# Define the RGB Images path
input_rgb = "2. TrainingPreparation/Enschede3D/rgb"

# Define the input txt file
input_list = "2. TrainingPreparation/Enschede3D/valid_list.txt"

# Define the output folder path
output_folder = "2. TrainingPreparation/Enschede3D/GT"
```

```python
listgt= open(input_list,'r').read().split()
print(listgt)
```

```python
for k in listgt:
    my_array = np.load(input_annot+'/'+str(k)+'.npy', allow_pickle=True, encoding='bytes').tolist()
    x_coords = []
    y_coords = []
    cp = []
    for key, value in my_array.items():
        x_coords.append(key[0])
        y_coords.append(key[1])
        cp.append(value)
    
    # Plot the points and lines
    # fig, ax = plt.subplots()
    # plt.gca().invert_yaxis()
    # Add the x and y coordinates to the plot
    
    my_img = image.imread(input_rgb+'/'+str(k)+'.jpg')
    plt.imshow(my_img)
    
    for i in range(len(my_array.keys())):
        # loop through all values (connected points) of this key (corner)
        for j in range(len(cp[i])):
            # draw lines
            plt.plot([x_coords[i],cp[i][j][0]],
                [y_coords[i],cp[i][j][1]],'co-')
            
    plt.scatter(x_coords, y_coords, c='red', zorder=3)
    
    plt.xlim(0, 255)
    plt.ylim(255, 0)

    plt.axis("off")
    plt.tight_layout()
    print('saving ' + str(k))
    plt.savefig(output_folder+'/'+str(k)+".png")
    plt.close()
    # break
```
