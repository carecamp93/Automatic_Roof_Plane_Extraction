```python
# Extract the x and y coordinates and the points
x_coords = []
y_coords = []
cp = []
```

```python
for key, value in my_array.items():
    x_coords.append(key[0])
    y_coords.append(key[1])
    cp.append(value)
```

```python
# Plot the points and lines
fig, ax = plt.subplots()
# plt.gca().invert_yaxis()
for i in range(len(my_array.keys())):
    ax.plot([x_coords[i]], 
            [y_coords[i]], 
            '-o')
```

```python
# Plot the points and lines
fig, ax = plt.subplots()
# plt.gca().invert_yaxis()
# Add the x and y coordinates to the plot
ax.scatter(x_coords, y_coords, color='red')
for i in range(len(my_array.keys())):
    # loop through all values (connected points) of this key (corner)
    for j in range(len(cp[i])):
        # draw lines
        ax.plot([x_coords[i],cp[i][j][0]],
                [y_coords[i],cp[i][j][1]],'ro-')
plt.xlim(0, 256)
plt.ylim(0, 256)
        
plt.show()
```
